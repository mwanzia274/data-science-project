```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the axes ranges
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
z = np.linspace(-1, 1, 100)

# Create a meshgrid from the X and Y values
X, Y = np.meshgrid(x, y)

# Calculate Z values for the grid
Z = np.sqrt(X**2 + Y**2)

# Plot the X, Y, and Z axes
ax.plot([0, 1], [0, 0], [0, 0], color='red', label='X-axis')
ax.plot([0, 0], [0, 1], [0, 0], color='green', label='Y-axis')
ax.plot([0, 0], [0, 0], [0, 1], color='blue', label='Z-axis')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the parameters of the circle
radius = 1
theta = np.linspace(0, 2*np.pi, 100)  # Angle values
x = radius * np.cos(theta)  # X coordinates of the circle points
y = radius * np.sin(theta)  # Y coordinates of the circle points
z = np.zeros_like(x)  # Z coordinates (all zeros in this case)

# Plot the circle
ax.plot(x, y, z, label='Circle in 3D')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create random data points
np.random.seed(0)
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a 3D scatter plot
ax.scatter(x, y, z, c='b', marker='o', label='Random Data')

# Add labels and a legend
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of X, Y values
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

# Calculate Z values for the surface
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a 3D surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create random data
x = np.arange(5)
y = np.arange(5)
X, Y = np.meshgrid(x, y)
Z = np.random.rand(5, 5)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a 3D bar plot
ax.bar3d(X.ravel(), Y.ravel(), np.zeros_like(Z).ravel(), 1, 1, Z.ravel(), shade=True)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of X, Y, Z vectors
X, Y, Z = np.meshgrid(np.linspace(-5, 5, 5),
                     np.linspace(-5, 5, 5),
                     np.linspace(-5, 5, 5))

# Create velocity vectors
U = np.sin(X) * np.cos(Y) * np.cos(Z)
V = -np.cos(X) * np.sin(Y) * np.cos(Z)
W = 0.5 * np.sin(Z)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a 3D quiver plot
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True, cmap='viridis')

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```
